### Дополнительно в срезах можно указывать шаг перебора элементов, согласно синтаксису:
#### список[start:stop:step]

```python
print(marks)
```

```python
marks[1:5:2]
```

```python
marks[:5:2]
```

```python
marks[::2]
```

```python
marks[::-1]
```

```python
marks[::-2]
```

### механизм срезов работает абсолютно также как и со строками, только здесь они применяются к спискам

```python
print(marks)
```

```python
marks[2:4] = ["хорошо", "удовлетворительно"]
print(marks)
```

```python
marks[::2] = [0, 0, 0]
print(marks)
```

```python
print(marks)
```

```python
marks += [3]
print(marks)
```

```python
marks[::2] = [0, 0, 0]
```

```python
marks[:5:2] = [0, 0, 0]
print(marks)
```

```python
marks[2:4] = 10, 20
print(marks)
```

### Сравнение списков

```python
[1, 2, 3] == [1, 2, 3]
```

```python
[10, 2, 3] > [1, 2, 3]
```

```python
[10, 2, 3] < [1, 2, 3]
```

```python
[1, 2, "abc"] > [1, 2, "abc"]
```

```python
[1, 2, 3] > [1, 2, "abc"]
```

#### произойдет ошибка, т.к. число 3 не может быть сравнено со строкой «abc»

```python
v = [1205, 1101, 1434, 1320, 923, 874]
print(v[:3])
```

```python
v = [1205, 1101, 1434, 1320, 923, 874]
print(v[-4:])
```

```python
c = ["Москва", "Ульяновск", "Самара", "Тверь", "Вологда", "Омск", "Уфа"]
print(c[::2])
```

```python
c = ["Москва", "Ульяновск", "Самара", "Тверь", "Вологда", "Омск", "Уфа"]
print(c[1::2])
```
