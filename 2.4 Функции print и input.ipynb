## 2.4 Функции print и input

* print() - вывод данных в консоль
* input() - ввод данных из стандартного входного потока (часто клавиатуры)

### Функция print()

```python
print(1)
```

```python
print(2,3)
```

```python
a = -6.84
print(a)
```

```python
print(a*2+3)
```

```python
print(abs(a*2+3))
```

* sep - разделитель между данными
* end - завершающий символ или строка

```python
a = -6.84
b = 7
c = 25.6
print(a, b, c)
```

```python
print(a, b, c, sep=" | ")
```

```python
print(a, sep=" | ")
```

```python
print(a, b, sep=" | ")
```

По умолчанию end = '\n' (перевод на следующую строку)

```python
print("Hello")
print("World")
```

```python
print("Hello", end=' ')
print("World")
```

```python
x = 5.76
y = -8
print("Координаты точки: x = ", x, "; y = ", y)
```

```python
print(f"Координаты точки: x = {x}, y = {y}")
```

## Функция input()

```python
a = input()
print(type(a))
```

```python
a = input()
b = abs(a)
```

```python
a = "54"
int(a)
```

```python
a = input()
a = int(a)
b = abs(a)
print(b)
```

```python
a = int(input())
```

```python
a = float(input())
b = abs(a)
print(b)
```

```python
a = float(input())
b = float(input())
print("Периметр:", 2 * (a + b))
```

```python
a = float(input("Введите длину прямоугольника: "))
b = float(input("Введите ширину прямоугольника: "))
print("Периметр:", 2 * (a + b))
```

```python
a, b = map(float, input("Введите две стороны прямоугольника:").split())
```

```python
a, b, c = map(int, input("Введите три целых числа: ").split())
print("Периметр треугольника:", a + b + c)
```

```python
a = 7
b = -4
c = 3
print(a,b,c)
```

```python
print(f"{a}\n{b}\n{c}")
```

```python
s1 = "Hello"
s2 = "Balakirev"
print(s1, end=" ")
print(s2)
```

```python
s1, s2 = map(str.strip, input().split())
print(f"Word 1: {s1}", f"Word 2: {s2}", sep=" | ")
```

```python
a, b = map(int, input().split())
print(a ** b)
```

```python
a, b = map(float, input().split())
print(a+b)
```

```python
X, Y = map(int, input().split())
print(X+Y+X*2+Y*4)
```

```python
x = float(input())
y = float(input())
print((x+y)*2)
```

```python
import math
print(round(math.pi, 3))
```

```python
print(f'{math.pi:.3f}')
```

```python
a = float(input())
print(f"Вы ввели число {a}")
```
