## 2.5 Арифметические подвиги

```python
a, b = map(float, input().split())
a, b = b, a
print(a)
print(b)
```

```python
count, total = map(float, input().split())
count += 2
total -= 0.3
```

```python
import math
a, b, c = map(float, input().split())
L = math.sqrt(a**2+b**2+c**2)
lenght = round(L,2)
lenght
```

```python
digit = int(input())
sum_digit = 0
while digit > 0:
    sum_digit += digit%10
    digit //= 10 
sum_digit
```

```python
#digit = int(input())
sum_digit = sum(map(int, input()))
sum_digit
```

```python
import math
a, b, c = map(int, input().split())
p = sum((a,b,c))/2
sq_tr = math.sqrt(p*(p-a)*(p-b)*(p-c))
sq_tr
```

```python
a, b, alpha = map(float, input().split())
length_c = math.sqrt(a**2+b**2-2*a*b*math.cos(alpha))
length_c
```

```python
d1, d2, alpha = map(int, input().split())
rad = alpha/180*math.pi
square = round((d1*d2)/2*math.sin(rad),1)
square
# square = round(d1 * d2 / 2 * math.sin(math.radians(alpha)), 1)
```

```python
import math
x0, y0, x1, y1 = map(float, input().split())
cos = (x0*x1 + y0*y1)/(math.sqrt(x0**2+y0**2)*math.sqrt(x1**2+y1**2))
alpha = math.acos(cos)
alpha = round(180/math.pi*alpha,1)
alpha
# alpha = round(math.degrees(math.acos(cos)), 1)
```
