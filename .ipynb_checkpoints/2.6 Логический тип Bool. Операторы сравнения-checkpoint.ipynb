| symbol | description                |
|--------|----------------------------|
|    <   | сравнение на меньше        |
|    >   | сравнение на больше        |
|   <=   | сравнение меньше или равно |
|   >=   | сравнение больше или равно |
|   ==   | сравнение на равенство     |
|   !=   | сравнение на неравенство   |

```python
from IPython.display import HTML, display

html_table = """
<table border="1">
    <thead>
        <tr>
            <th>Символ</th>
            <th>Описание</th>
        </tr>
    </thead>
    <tbody>
        <tr><td>&lt;</td><td>сравнение на меньше</td></tr>
        <tr><td>&gt;</td><td>сравнение на больше</td></tr>
        <tr><td>&lt;=</td><td>сравнение меньше или равно</td></tr>
        <tr><td>&gt;=</td><td>сравнение больше или равно</td></tr>
        <tr><td>==</td><td>сравнение на равенство</td></tr>
        <tr><td>!=</td><td>сравнение на неравенство</td></tr>
    </tbody>
</table>
"""

display(HTML(html_table))
```

### Операторы not, and, or и их приоритеты

```python
y = 1.85
y >= -2 and y <= 5
```

```python
-2 <= y <= 5
```

```python
y < -2 or y > 5
```

```python
x % 2 != 0 and x % 3 != 0
```

```python
not(x % 2 == 0 or x % 3 == 0)
```

| symbol | priority |
|--------|----------|
|   or   |    low   |
|  and   |  average |
|  not   |   high   |
