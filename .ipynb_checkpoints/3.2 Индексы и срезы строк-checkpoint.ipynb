```python
str = input()
print(str[0]+str[-1])
```

```python
# s = input()
# print(s[len(s)-3:])
print(input()[-3:])
```

```python
s = input()
print(s[1::2])
```

```python
s1 = input()
s2 = input()
s1 = s1[::2]
s2 = s2[1::2]
print(s1 + " " + s2)
```
