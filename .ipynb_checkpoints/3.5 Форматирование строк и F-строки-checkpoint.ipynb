## 3.5 Форматирование строк и F-строки

```python
age = 18
name = "Антон"
```

```python
print("Меня зовут " + name + ", мне " + str(age) + " и я люблю язык Python.")
```

#### str.format(*args)

```python
"Меня зовут {0}, мне {1} и я люблю язык Python.".format(name, age)
```

```python
msg = "Меня зовут {0}, мне {1} и я люблю язык Python.".format(name, age)
print(msg)
```

```python
"Меня зовут {0}, мне {1} и я люблю язык {0} Python.".format(name, age)
```

### format(fio=name, old=age)

```python
"Меня зовут {fio}, мне {old} и я люблю язык Python. {fio}".format(fio=name, old=age)
```

```python
"Меня зовут {name}, мне {old} и я люблю язык Python. {fio}".format(fio=name, old=age)
```

```python
"Меня зовут {fio}, мне {old} и я люблю язык Python. {fio}".format(old=age, fio=name)
```

#### f - строки

```python
print(f"Меня зовут {name}, мне {age} и я люблю язык Python.")
```

```python
print(f"Меня зовут {name.upper()}, мне {age * 2} и я люблю язык Python.")
```

```python
print(f"Меня зовут {len(name)}, мне {14*2} и я люблю язык Python.")
```

```python
name = input()
surname = input()
age = input()
res = "Уважаемый {0} {1}! Поздравляем Вас с {2}-летием!".format(name,surname,age)
print(res)
```

```python
w, d, h = input().split()
print("Габариты: {width} x {depth} x {height}".format(width=w, depth=d, height=h))
# print(f"Габариты: {input().replace(' ', ' x ')}")
# print("Габариты: {} x {} x {}".format(*input().split()))
```

```python
a, b = map(int,input().split())
print(f"{min(a,b)} {max(a,b)}")
```

```python
city = input()
street = input()
number_home = input()
number_flat = input()
print(f"г. {city}, ул. {street}, д. {number_home}, кв. {number_flat}")
# print (f"г. {input()}, ул. {input()}, д. {input()}, кв. {input()}")
```

```python
print('г. {}, ул. {}, д. {}, кв. {}'.format(*[input() for i in '♥♥♥♥']))
```

```python
dollar_rate = float(input().replace(',', '.'))
num_rub = int(input())
dollars = int(num_rub / dollar_rate)
print(f"Вы можете получить {dollars}$ за {num_rub} рублей по курсу {dollar_rate}")
```
