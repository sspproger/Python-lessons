## 3.9 Вложенные списки

```python
line = [1, 7, 6, 11, 3]
print(line)
```

```python
img = [[1, 7, 6, 11, 3], [1, 7, 6, 11, 3], [1, 7, 6, 11, 3], [1, 7, 6, 11, 3], [1, 7, 6, 11, 3]]
print(img)
```

```python
img = [line[:], line[:], line[:], line[:], line[:]]
print(img)
```

```python
print(img[0])
```

```python
print(img[0][1])
```

#### Мы сформировали новый объект – список из нулей, связали с ним вторую ссылку главного списка, а прежний список был автоматически удален сборщиком мусора.

```python
img[1] = [0, 0, 0, 0, 0]
print(img)
```

```python
img[1] = [0] * 5
print(img)
```

#### Чтобы изменить значения уже существующего вложенного списка, то нужно обратиться к его элементам, например, через механизм срезов:

```python
print(img)
```

```python
img[2][:] = [1] * 5
print(img)
```
