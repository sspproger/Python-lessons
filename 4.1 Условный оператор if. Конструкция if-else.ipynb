## 4.1 Условный оператор if. Конструкция if-else

```python
x = -4
if x < 0:
    x = -x
print(x)
```

```python
a = float(input("a: "))
b = float(input("b: "))
if a < b:
    a, b = b, a
print(a, b)
```

```python
x = int(input())
if x >= -4 and x <= 10:
    print("x в диапазоне [-4; 10]")
```

```python
x = int(input())
if -4 <= x <= 10:
    print("x в диапазоне [-4; 10]")
```

```python
x = int(input())
if x:
    print("x = True")
```

```python
x = int(input())
if x:
    print("x = True")
```

```python
if True:
    print("True")
```

```python
marks = [3, 3, 4, 2, 4]
if 2 in marks:
    print("Студент будет отчислен")
```

```python
marks = [3, 3, 4, 5, 4]
if 2 in marks:
    print("Студент будет отчислен")
else:
    print("студент успешно сдал сессию")
```

```python
x = int(input())
if x < 0:
    print("x - отрицательное число")
else:
    print("x - положительное число")
```

```python
x = int(input())
if x % 2 == 0:
    print("x - четное число")
else:
    print("x - нечетное число")
```

```python
a, b = map(float, input().split())
if a > b:
    print(a)
else:
    print(b)
```

```python
word = input().lower()
rev = word[::-1]
if word == rev:
    print("ДА")
else: 
    print("НЕТ")
```

```python
m, n = map(int, input().split())
if m % n == 0:
    print(m // n)
else:
    print(f"{m} на {n} нацело не делится")
```

```python
a, b, c = map(int, input().split())
if a**2 + b**2 == c**2:
    print("ДА")
else:
    print("НЕТ")
```

```python
num = input()
if num[-1] == '7':
    print("ДА")
else:
    print("НЕТ")

# num = input()[-1]
# print('ДА' if num == '7' else 'НЕТ')
```

```python
word = input().lower()
if 't' in word and 'h' in word and 'o' in word:
    print("ДА")
else:
    print("НЕТ")
```

```python
cities = input().split()
if "Москва" in cities:
    cities.remove("Москва")
print(*cities)
```

```python
a, b, c, d = map(int, input().split())
if (a >= c + 2 and b >= d + 2) or (a >= d + 2 and b >= c + 2):
    print("ДА")
else:
    print("НЕТ")
```

```python
num = list(map(int, input()))
sum1 = sum(num[:3])
sum2 = sum(num[3:])
if sum1 == sum2:
    print("ДА")
else:
    print("НЕТ")
```
